```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

# Set your API key
api_key = 'AIzaSyChdbFBe2pT22iRdKSW36_AjB4H6D5o3qo'

# Create a YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

def get_channel_videos(channel_id):
    videos = []
    next_page_token = None
    
    try:
        while True:
            request = youtube.search().list(
                part='snippet',
                channelId=channel_id,
                maxResults=50,
                order='date',
                pageToken=next_page_token
            )
            
            response = request.execute()
            
            videos += response['items']
            
            next_page_token = response.get('nextPageToken')
            
            if not next_page_token:
                break
    except HttpError as e:
        print('An HTTP error occurred:')
        print(e.content)
        return None
    
    return videos

def get_video_stats(video_id):
    try:
        request = youtube.videos().list(
            part="snippet,statistics,contentDetails",
            id=video_id
        )
        response = request.execute()
        video = response['items'][0]
        
        video_data = {
            "Video ID": video_id,
            "Title": video["snippet"]["title"],
            "Published At": video["snippet"]["publishedAt"],
            "Views": video["statistics"].get("viewCount", 0),
            "Likes": video["statistics"].get("likeCount", 0),
            "Dislikes": video["statistics"].get("dislikeCount", 0),
            "Comments": video["statistics"].get("commentCount", 0),
            "Duration": video["contentDetails"]["duration"],
            "Tags": video["snippet"].get("tags", []),
            "Category": video["snippet"].get("categoryId", ""),
            "Description": video["snippet"].get("description", ""),
            "Channel Name": get_channel_name(video["snippet"]["channelId"]),
            "Thumbnail URL": video["snippet"]["thumbnails"]["default"]["url"],
            "Video URL": f"https://www.youtube.com/watch?v={video_id}",
            # You can add more fields here based on your requirements
        }
        
        return video_data
    except HttpError as e:
        print("An error occurred:", e)
        return None

def get_channel_subscribers(channel_id):
    try:
        request = youtube.channels().list(
            part='statistics',
            id=channel_id
        )
        response = request.execute()
        return response['items'][0]['statistics']['subscriberCount']
    except HttpError as e:
        print("An error occurred:", e)
        return None

def get_channel_name(channel_id):
    try:
        request = youtube.channels().list(
            part='snippet',
            id=channel_id
        )
        response = request.execute()
        return response['items'][0]['snippet']['title']
    except HttpError as e:
        print("An error occurred:", e)
        return None

# Sony Entertainment Television India channel ID
channel_id = 'UCpEhnqL0y41EpW2TvWAHD7Q'

# Get videos from the channel
channel_videos = get_channel_videos(channel_id)

if channel_videos is not None:
    for video in channel_videos:
        if 'id' in video and 'videoId' in video['id']:
            video_id = video['id']['videoId']
            video_stats = get_video_stats(video_id)
            if video_stats is not None:
                print(video_stats)
        else:
            print("Skipping video without 'videoId' key:", video)
```

```python
# So now you can see in data that data of column "duration" is not in proper format that person can't able to understand so,
# we create a function to set it in proper format that we can understant it.

def convert_duration(duration_str):
    if duration_str.startswith('PT'):
        duration_str = duration_str[2:]  # Remove "PT" prefix
        hours = minutes = seconds = 0

        if 'H' in duration_str:
            hours = int(duration_str.split('H')[0])
            duration_str = duration_str.split('H')[1]

        if 'M' in duration_str:
            minutes = int(duration_str.split('M')[0])
            duration_str = duration_str.split('M')[1]

        if 'S' in duration_str:
            seconds = int(duration_str.split('S')[0])

        return '{:02d}-{:02d}-{:02d}'.format(hours, minutes, seconds)
    else:
        return 'Invalid Format'

# Apply the function to the Duration column
data['Duration'] = data['Duration'].apply(convert_duration)
```

```python
# we are changing the column name

data.rename(columns={'Duration': 'Duration HH-MM-SS'}, inplace=True)
```
